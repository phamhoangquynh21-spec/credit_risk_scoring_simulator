# 01 — Exploratory Data Analysis

Credit Risk Scoring Simulator. Explores the (synthetic, UCI-structured)
dataset: distributions, target imbalance, correlations.

> Run `python -m src.generate_data` first if `data/raw/` is empty.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from src import config
pd.set_option('display.max_columns', 40)
```

```python
from src.preprocessing import load_data, clean_data, engineer_features
raw = load_data()
print(raw.shape)
raw.head()
```

## Cleaning & feature engineering

```python
df = engineer_features(clean_data(raw))
print(df.shape)
df[['limit_bal','age','credit_utilization','months_delayed_count'] ].describe()
```

## Target balance (class imbalance)

```python
rate = df[config.TARGET].mean()
print(f'Default rate: {rate:.1%}')
df[config.TARGET].value_counts().plot(kind='bar', color=['#2ca02c','#d62728']);
plt.title('Default vs No-default'); plt.show()
```

## Distributions of key drivers

```python
fig, axes = plt.subplots(1, 3, figsize=(14,3))
df['limit_bal'].plot(kind='hist', bins=40, ax=axes[0], title='Credit limit')
df['age'].plot(kind='hist', bins=40, ax=axes[1], title='Age')
df['credit_utilization'].clip(0,2).plot(kind='hist', bins=40, ax=axes[2], title='Credit utilisation')
plt.tight_layout(); plt.show()
```

## Default rate by repayment status (PAY_0)

```python
df.groupby('pay_0')[config.TARGET].mean().plot(kind='bar', color='#ff7f0e');
plt.ylabel('Default rate'); plt.title('Default rate by most-recent repayment status'); plt.show()
```

## Correlation with the target

```python
corr = df.corr(numeric_only=True)[config.TARGET].drop(config.TARGET).sort_values()
corr.plot(kind='barh', figsize=(6,7)); plt.title('Correlation with default'); plt.show()
```

**Takeaways:** recent repayment delays (`pay_0`, `months_delayed_count`) and
credit utilisation are the strongest correlates of default — consistent with
domain expectations and with the feature importances in notebook 02.
